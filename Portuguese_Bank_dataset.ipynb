# Optimization using campign

```python
sum(data['y_yes'])/sum(data['campaign'])
# Total Conversion ratio
```

```python
# Now let's see efficiancy on every additional call
print("Nth Call \t Efficiency")
for i in range(1,30):
    goo = sum(data.loc[data['campaign']==i]['y_yes']) / float(data.loc[data['campaign'] >= i].shape[0])
    print (str((i))+" \t\t "+str(goo))
```

Here we observe that after 6 calls efficiency drops below 1.6%, so maximum no of calls for efficiency is 6.

```python
# Calculate how many calls were made in total
total_calls = sum(data['campaign'])
print(total_calls)
```

```python
# Calculate how many calls were made after the 6th call
extra_calls = sum(data[data['campaign']>6]['campaign']) - 6*data[data['campaign']>6].shape[0]
print(extra_calls)
```

```python
# Calculate reduction in marketing cost
reduction=100*extra_calls/total_calls
print(reduction)
```

```python
total_sales=float(data[data['y_yes']==1].shape[0])
print(total_sales)
```

```python
less_costly_sales=float(data[(data['campaign'] <= 6) & (data['y_yes']==1)].shape[0])
print(less_costly_sales)
```

```python
sales_percent=100*less_costly_sales/total_sales
print(sales_percent)
```

# Hence a reduction of about 11.4% in marketing cost can be achieved while maintaining 97.6% sales if any person is called a maximum of 6 times.
